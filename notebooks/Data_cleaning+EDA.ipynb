```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header=1)
dataset.head()
```

# Data Cleaning

```python
dataset.info()
```

#Data Cleaning

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122, "Region"]=0
dataset.loc[122:, "Region"]=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.isnull().sum()
```

```python
#removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
#fix spaces in column names
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

```python
df.head(2)
```

```python
#change the required columns as integer data type

df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]= df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
#changing other columns to float datatype

df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']]= df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
#Saving cleaned dataset
df.to_csv('Algerian_forest_fires_dataset_Cleaned.csv',index=False)
```

# Exploratory Data Analysis

```python
#Dropping day, month, year as unnecessary
df_copy = df.drop(['day', 'month', 'year'], axis=1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
#Encoding Classes column
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
```

```python
df_copy.head()
```

```python
#categories in classes
df_copy['Classes'].value_counts()
```

```python
#ploting density plot for all features
df_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
#Percentage for pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
# Plotting pie chart
classlabels = ['Fire',' Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')
plt.title('Pie chart of classes')
plt.show()
```

```python
#Correlation
df_copy.corr()
```

```python
#Box plots
sns.boxplot(df['FWI'], color='grey')
plt.show()
```
